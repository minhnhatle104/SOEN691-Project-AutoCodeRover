```python
from datasets import load_dataset
import pandas as pd

# Load dataset
ds = load_dataset("Daoguang/Multi-SWE-bench")

# Convert to Pandas DataFrame
df = pd.DataFrame(ds['java_verified'])  # Change 'train' to the appropriate split if needed

# Save as Parquet file
parquet_filename = "multi_swe_bench.parquet"
df.to_parquet(parquet_filename, engine='pyarrow')

# Download file
from google.colab import files
files.download(parquet_filename)
```
